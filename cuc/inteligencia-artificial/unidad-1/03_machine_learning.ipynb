## Librerías necesarias

```python
# Importar módulos necesarios
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
```

## DataFrame de load_iris

```python
# Crear DataFrame de load_iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
iris_df['species'] = iris['target']
iris_df.head(10)
```

## Gráfico del DataFrame

```python
# Generar gráfico del DataFrame
plt.figure(figsize=(10, 6))
sns.countplot(x='species', data=iris_df, hue='species', palette='Paired', legend=False)
plt.xlabel('Especies (0: setosa, 1: versicolor, 2: virginica)')
plt.ylabel('Frecuencia')
plt.title('Distribución de las especies de Iris')
plt.grid(axis='y', linestyle='--')
plt.show()
```

## Dividir datos para entrenamientos

```python
# Dividimos datos para pruebas y entrenamientos
X = iris_df.drop(columns='species')
y = iris_df['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Entrenamos el modelo
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy * 100:.2f}%')
```

## Visualizar importancia de las características

```python
# Visualizar importancias de las caracterísitcas (Seaborn)
feature_importances = pd.DataFrame({
    'feature': iris['feature_names'],
    'importance': clf.feature_importances_
})

sns.barplot(x='importance', y='feature', data=feature_importances, palette='RdBu', hue='feature', legend=False)
plt.xlabel('Importancia')
plt.ylabel('Características')
plt.title('Importancia de las Características del Modelo')
plt.grid(axis='x', linestyle='--')
plt.show()
```
