## Parte 1

**Descripción:**

Crear un análisis simple de las ventas de un pequeño negocio, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados
con Matplotlib.

- [x] Creación DataFrame
- [x] Operaciones de NumPy
- [x] Visualización con Matplotlib


### Módulos necesarios

```python
# Módulos necesarios
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### DataFrame con arrays de NumPy

```python
# Crear arrays de NumPy
meses_np = np.array(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                     'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre',
                     'Diciembre'])
ventas_np = np.array([25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90])

# Crear un DataFrame a partir de los arrays de NumPy
data_np = {'mes': meses_np, 'ventas': ventas_np}
df = pd.DataFrame(data_np)

# Mostrar el DataFrame
print("DataFrame creado con arrays de NumPy:")
print(df)
```

### Analizar datos con Pandas

```python
# Realizar análisis estadístico descriptivo de la columna 'ventas'
descripcion_ventas = df['ventas'].describe()

# Mostrar los resultados
print("Análisis estadístico descriptivo de las ventas:")
print(descripcion_ventas)
```

### Analizar datos con NumPy

```python
# 1. Suma total de ventas usando np.sum()
total_ventas = np.sum(df['ventas'])
print(f"Suma total de ventas (usando np.sum): {total_ventas}")

# 2. Promedio de ventas usando np.mean()
promedio_ventas = np.mean(df['ventas'])
print(f"Promedio de ventas (usando np.mean): {promedio_ventas}")

# 3. Venta máxima usando np.max()
venta_maxima = np.max(df['ventas'])
print(f"Venta máxima (usando np.max): {venta_maxima}")

# 4. Venta mínima usando np.min()
venta_minima = np.min(df['ventas'])
print(f"Venta mínima (usando np.min): {venta_minima}")

# 5. Meses con ventas superiores al promedio usando enmascaramiento de NumPy
meses_altas_ventas = df['mes'][df['ventas'] > np.mean(df['ventas'])]
print(f"Meses con ventas superiores al promedio:\n{meses_altas_ventas.tolist()}")
```

### Gráfico con Matplotlib

```python
# Generación del gráfico con el DataFrame
plt.plot(df['mes'], df['ventas'], marker='+', linestyle='-', color='teal')
plt.xlabel('mes')
plt.ylabel('ventas')
plt.title('mes vs ventas')
plt.xticks(rotation=90)
plt.show()
```

Realizar un análisis simple de las calificaciones de estudiantes en diferentes asignaturas, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados con Matplotlib.

### Módulos necesarios

```python
# Módulos necesarios
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### DataFrame con arrays de NumPy

```python
# Crear arrays de NumPy
estudiantes_np = np.array(["Alice", "Bob", "Charlie", "David", "Eva", "Frank",
                           "Grace", "Hannah", "Ian", "Jack"])
matematicas_np = np.array([85, 92, 78, 90, 88, 76, 95, 89, 82, 94])
ciencias_np = np.array([90, 85, 80, 88, 92, 75, 95, 87, 81, 93])
historia_np = np.array([78, 82, 88, 90, 85, 80, 92, 84, 79, 91])

# Crear un DataFrame a partir de los arrays de NumPy
data_np = {
    'Estudiantes': estudiantes_np,
    'Matemáticas': matematicas_np,
    'Ciencias': ciencias_np,
    'Historia': historia_np,
}
df = pd.DataFrame(data_np)

# Mostrar el DataFrame
print("DataFrame creado con arrays de NumPy:")
print(df)
```

### Analizar datos con Pandas

```python
# # Realizar análisis estadístico descriptivo de asignaturas:
print("Análisis estadístico descriptivo de las asignaturas:")
df.describe()
```

### Analizar datos con NumPy

```python
# 1. Notas máximas asignaturas:
notas_maximas = df.iloc[:, 1:].max()
print(f"Notas máximas por asignatura:\n{notas_maximas}\n")

# 2. Notas mínimas asignaturas:
notas_minimas = df.iloc[:, 1:].min()
print(f"Notas mínimas por asignatura:\n{notas_minimas}\n")

# 3. Promedios de notas asignaturas:
promedio_notas = df.iloc[:, 1:].mean()
print(f"Promedios de notas por asignatura:\n{promedio_notas}\n")

# 4. Promedio de todas las asignaturas:
promedio_total = df.iloc[:, 1:].mean().mean()
print(f"Promedio de todas las asignaturas: {promedio_total}\n")

# 5. Promedio de notas por estudiante:
df['Promedio'] = df.iloc[:, 1:].mean(axis=1)
print("DataFrame con el promedio de notas por estudiante:")
print(df)

# 6. Estudiante promedio mayor:
estudiante_promedio_mayor = df.loc[df['Promedio'].idxmax(), 'Estudiantes']
print(f"\nEstudiante con el promedio más alto: {estudiante_promedio_mayor}")

# 7. Estudiante promedio menor:
estudiante_promedio_menor = df.loc[df['Promedio'].idxmin(), 'Estudiantes']
print(f"Estudiante con el promedio más bajo: {estudiante_promedio_menor}")

# 8. Promedio de promedios de estudiantes:
promedio_promedios = df['Promedio'].mean()
print(f"\nPromedio de promedios de estudiantes: {promedio_promedios}")
```

### Gráfico con Matplotlib

```python
# Generación del gráfico con el DataFrame

# Promedio notas por estudiantes
plt.figure(figsize=(10, 6))
plt.bar(df['Estudiantes'], df['Promedio'], color='teal')
plt.xlabel('Estudiantes')
plt.ylabel('Promedio')
plt.title('Promedio de Notas por Estudiante')
plt.grid(axis='y', linestyle='--') # Añadir cuadrícula en el eje y

# Añadir líneas horizontales desde la parte superior de cada barra al eje y
for i, promedio in enumerate(df['Promedio']):
    plt.hlines(promedio, -0.4 + i, i + 0.4, color='gray', linestyle='-', linewidth=0.7) # Ajustar la posición x de las líneas

# Establecer los ticks del eje y para que vayan de 10 en 10
plt.yticks(np.arange(0, df['Promedio'].max() + 10, 10))

plt.show()
```

```python
# Generación del gráfico con el DataFrame

# Promedio notas por asignatura
promedio_por_asignatura = df.iloc[:, 1:4].mean()
asignaturas = promedio_por_asignatura.index

plt.figure(figsize=(10, 6))
plt.bar(asignaturas, promedio_por_asignatura, color='teal')
plt.xlabel('Asignaturas')
plt.ylabel('Promedio')
plt.title('Promedio de Notas por Asignatura')
plt.grid(axis='y', linestyle='--') # Añadir cuadrícula en el eje y

# Añadir líneas horizontales desde la parte superior de cada barra al eje y
for i, promedio in enumerate(promedio_por_asignatura):
    plt.hlines(promedio, -0.4 + i, i + 0.4, color='gray', linestyle='-', linewidth=0.7) # Ajustar la posición x de las líneas

# Establecer los ticks del eje y para que vayan de 10 en 10
plt.yticks(np.arange(0, promedio_por_asignatura.max() + 10, 10))


plt.show()
```
